# Drill: Present the Same Data Multiple Ways - Kristofer Schobert

## Using Data on Olympic Athletes

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import ttest_ind
import seaborn as sns
#%matplotlib inline

dataframe = pd.read_csv('athlete_events.csv')
df = pd.DataFrame(dataframe)
df.head(7)
```

# One Continuous Variable Plotted Four Ways

```python
sns.set(style="whitegrid", palette="pastel")
ax = sns.boxplot(x=df[df["Year"] == 2016][df[df["Year"] == 2016]["Event"] == "Athletics Women's 100 metres"]['Height'])
plt.xlabel('Height (cm)')
plt.title('Boxplot: Women\'s 100 Meter Sprinters at Rio 2016')
plt.show()


sns.set(style="whitegrid")
ax = sns.swarmplot(x=df[df["Year"] == 2016][df[df["Year"] == 2016]["Event"] == "Athletics Women's 100 metres"]['Height'])
plt.xlabel('Height (cm)')
plt.title('Swarmplot: Women\'s 100 Meter Sprinters at Rio 2016')
plt.show()



sns.set(style="whitegrid")
ax = sns.violinplot(x=df[df["Year"] == 2016][df[df["Year"] == 2016]["Event"] == "Athletics Women's 100 metres"]['Height'])
plt.xlabel('Height (cm)')
plt.title('Violinplot: Women\'s 100 Meter Sprinters at Rio 2016')
plt.show()


sns.set(style="whitegrid")
ax = sns.stripplot(x=df[df["Year"] == 2016][df[df["Year"] == 2016]["Event"] == "Athletics Women's 100 metres"]['Height'])
#this is like a swarmplot, but the points are randomly distributed along the y-direction and thus they can overlap
plt.xlabel('Height (cm)')
plt.title('Stripplot: Women\'s 100 Meter Sprinters at Rio 2016 \n (this is like a swarmplot, but the points are randomly distributed \n in the y-direction and thus they can overlap)')
plt.show()
```

# Two Continuous Variables Plotted Three Ways
## The Height and Weight of the Snowboarding Men's Halfpipe Olympians at the 2014 Winter Games in Sochi, Russia

```python
g = sns.scatterplot(y='Height', 
               x='Weight',
               data=df.loc[(df['Year']==2014)&(df['Event']=="Snowboarding Men's Halfpipe")])
               #fit_reg=True, # The regression line also includes a 95% confidence envelope.
               #scatter_kws={'alpha':0.4})
plt.ylabel("Height (cm)")
plt.xlabel("Weight (kg)")
plt.title('Scatterplot: Height vs Weight \n for Snowboarding Men\'s Halfpipe Olympians in 2014')
plt.show()


g = sns.lmplot(y='Height', 
               x='Weight',
               data=df.loc[(df['Year']==2014)&(df['Event']=="Snowboarding Men's Halfpipe")],
               fit_reg=True) # The regression line also includes a 95% confidence envelope.
               #scatter_kws={'alpha':0.4})
g.set_ylabels("Height (cm)")
g.set_xlabels("Weight (kg)")
plt.title('lmplot: Height vs Weight \n for Snowboarding Men\'s Halfpipe Olympians in 2014')
plt.show()

g = sns.lmplot(y='Height', 
               x='Weight',
               data=df.loc[(df['Year']==2014)&(df['Event']=="Snowboarding Men's Halfpipe")],
               fit_reg=True, # The regression line also includes a 95% confidence envelope.
               #scatter_kws={'alpha':0.4},
               lowess=True)
            
plt.ylabel("Height (cm)")
plt.xlabel("Weight (kg)")
plt.title('Scatterplot with adjusting regression line: Height vs Weight \n for Snowboarding Men\'s Halfpipe Olympians in 2014')
plt.show()
```

# One Continuous Variable and One Catagorial Variable Plotted Six Ways
## The Ages of Men and Women Competeing in the Figure Skating Pairs Competition at the Winter 1992 Games of Albertville, France

```python

sns.set(style="ticks", palette="pastel")
sns.boxplot(x="Age", 
            y="Sex",
            #hue="Sex", 
            palette=["m","g"],
            data=df.loc[(df['Year'] == 1992)&(df['Event'] == 'Figure Skating Mixed Pairs')])
sns.despine(offset=0, trim=False)
plt.title('Boxplot: Figure Skating Mixed Pairs 1992')
plt.show()




sns.set(style="ticks")
sns.swarmplot(x="Age", 
            y="Sex",
            #hue="Sex", 
            palette=["m","g"],
            data=df.loc[(df['Year'] == 1992)&(df['Event'] == 'Figure Skating Mixed Pairs')])
sns.despine(offset=0, trim=False)
plt.title('Swarmplot: Figure Skating Mixed Pairs 1992')
plt.show()



sns.set(style="ticks", palette="pastel")
sns.barplot(x="Age", 
            y="Sex",
            #hue="Sex", 
            palette=["m","g"],
            data=df.loc[(df['Year'] == 1992)&(df['Event'] == 'Figure Skating Mixed Pairs')])
sns.despine(offset=0, trim=False)
plt.title('Barplot: Figure Skating Mixed Pairs 1992 \n (this is not a great plot since the age of zero is not of significance)')
plt.show()




```

```python

sns.set(style="ticks")
sns.violinplot(x="Age", 
            y="Sex",
            #hue="Sex", 
            palette=["m","g"],
            data=df.loc[(df['Year'] == 1992)&(df['Event'] == 'Figure Skating Mixed Pairs')])
sns.despine(offset=0, trim=False)
plt.title('Violinplot: Figure Skating Mixed Pairs 1992')
plt.show()



sns.set(style="ticks")
sns.pointplot(y="Age", 
            x="Sex",
            #hue="Year", 
            palette=["m","g"],
            data=df.loc[(df['Year'] == 1992)&(df['Event'] == 'Figure Skating Mixed Pairs')])
sns.despine(offset=0, trim=False)
plt.title('Pointplot: Figure Skating Mixed Pairs 1992')
plt.show()




sns.set(style="ticks")
sns.swarmplot(y="Age", 
            x="Year",
            hue="Sex", 
            palette=["m","g"],
            data=df.loc[(df['Year'] == 1992)&(df['Event'] == 'Figure Skating Mixed Pairs')])
sns.despine(offset=0, trim=False)
plt.title('Swarmtplot: Figure Skating Mixed Pairs 1992')
plt.show()


```

# A BoxenPlot for Our Three Continuous Variables Catagorized by Each Olympic Games

```python
plt.figure(figsize=(20,10))
sns.set(style="whitegrid")
sns.boxenplot(x="Year", y="Age",
              color="b",
              scale="linear", data=df)


plt.figure(figsize=(20,10))
sns.set(style="whitegrid")
sns.boxenplot(x="Year", y="Height",
              color="b",
              scale="linear", data=df)
plt.ylabel('Height (cm)')


plt.figure(figsize=(20,10))
sns.set(style="whitegrid")
sns.boxenplot(x="Year", y="Weight",
              color="b",
              scale="linear", data=df)
plt.ylabel('Weight (kg)')
plt.show()
```
